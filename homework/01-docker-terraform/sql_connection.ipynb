```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
green_trip_iter = pd.read_csv('../datasets/green_tripdata_2019-09.csv',iterator=True,chunksize=100000)
green_trip_chunk = next(green_trip_iter)
```

```python
# green_trip_chunk.head(n=0).to_sql(name="green_trip",con=engine,if_exists="replace")
```

```python
table_create_query = pd.io.sql.get_schema(green_trip_chunk,name = 'green_trip',con=engine)
print(table_create_query)
```

```python
table_create_query = """
DROP TABLE IF EXISTS green_trip;
CREATE TABLE green_trip (
	"VendorID" BIGINT, 
	lpep_pickup_datetime TIMESTAMP, 
	lpep_dropoff_datetime TIMESTAMP , 
	store_and_fwd_flag TEXT, 
	"RatecodeID" BIGINT, 
	"PULocationID" BIGINT, 
	"DOLocationID" BIGINT, 
	passenger_count BIGINT, 
	trip_distance FLOAT(53), 
	fare_amount FLOAT(53), 
	extra FLOAT(53), 
	mta_tax FLOAT(53), 
	tip_amount FLOAT(53), 
	tolls_amount FLOAT(53), 
	ehail_fee FLOAT(53), 
	improvement_surcharge FLOAT(53), 
	total_amount FLOAT(53), 
	payment_type BIGINT, 
	trip_type BIGINT, 
	congestion_surcharge FLOAT(53)
)
"""
```

```python
try:
    pd.read_sql(table_create_query,con=engine)
except sqlalchemy.exc.ResourceClosedError:
    print("Created table")
```

```python
query = """ 
SELECT * FROM information_schema.columns WHERE table_name = 'green_trip';
"""

pd.read_sql(query,con=engine)
```

```python
from time import time
while True:
    try:
        t_start = time()
        # green_trip_chunk["lpep_pickup_datetime"] = pd.to_datetime(green_trip_chunk["lpep_pickup_datetime"])
        # green_trip_chunk["lpep_dropoff_datetime"] = pd.to_datetime(green_trip_chunk["lpep_dropoff_datetime"])
        green_trip_chunk.to_sql("green_trip",con=engine,if_exists="append",index=False)
        t_end = time()
        green_trip_chunk = next(green_trip_iter)
        print(f"Inserted another chunk %.3f seconds ({t_end - t_start})" % (t_end-t_start))
    except StopIteration:
        print("Reached end")
        break
```

```python

query = """
SELECT COUNT(*)
FROM green_trip
"""
pd.read_sql(query,con=engine)
```

## Questions

```python

query = """
SELECT 
    COUNT(*)
FROM 
    green_trip
WHERE 
    lpep_pickup_datetime::date >= '2019-09-18'::date
AND 
    lpep_dropoff_datetime::date = '2019-09-18'
"""
pd.read_sql(query,con=engine)
```

```python
query = """ 
SELECT 
    lpep_pickup_datetime 
FROM 
    green_trip 
WHERE 
    trip_distance = (
        SELECT 
            MAX(trip_distance) 
        FROM 
            green_trip
        )
"""

pd.read_sql(query,con=engine)
# SELECT lpep_pickup_datetime FROM green_trip HAVING MAX(trip_distance)
```

```python
query = """ 
SELECT 
    taxi_zone_lookup."Borough",
    SUM(total_amount)
FROM 
    green_trip
INNER JOIN taxi_zone_lookup
    ON green_trip."PULocationID" = taxi_zone_lookup."LocationID"
WHERE 
    green_trip.lpep_pickup_datetime::date='2019-09-18'
AND
    taxi_zone_lookup."Borough" != 'Unknown'
GROUP BY
    taxi_zone_lookup."Borough"
HAVING
    SUM(total_amount) > 50000
ORDER BY 
    sum DESC
"""

# WHERE
#     lpep_pickup_datetime::date='2019-09-18'
pd.read_sql(query,con=engine)
```

```python
query = """ 
    SELECT 
        tip_amount,
        green_trip."DOLocationID",
        taxi_zone_lookup."LocationID",
        "Zone"
    FROM
        green_trip
        INNER JOIN taxi_zone_lookup
            ON green_trip."DOLocationID" = taxi_zone_lookup."LocationID"
        WHERE 
        tip_amount = ( 
            SELECT 
                MAX(tip_amount)
            FROM
                green_trip
            INNER JOIN taxi_zone_lookup
                ON green_trip."PULocationID" = taxi_zone_lookup."LocationID"
            WHERE
                "Zone"='Astoria'
            AND
                green_trip.lpep_pickup_datetime::date>='2019-09-01'
            AND
                green_trip.lpep_pickup_datetime::date<='2019-09-30'
        )

"""

# WHERE
#     lpep_pickup_datetime::date='2019-09-18'
pd.read_sql(query,con=engine)
```
